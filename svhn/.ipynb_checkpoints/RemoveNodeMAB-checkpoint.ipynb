# Compare the accuracy of the models

```python
AccBefore = np.load('/home/salm/Documents/svhn/AccuracyBEFOREPrune.npy')
ucb1 = np.load('/home/salm/Documents/svhn/IrisUCB1/AccuracyAftrerPrune.npy')
EpsilonGreedy = np.load('/home/salm/Documents/svhn/IrisEpsilonGreedy/AccuracyAftrerPrune.npy')
AnnealingEpsilonGreedy = np.load('/home/salm/Documents/svhn/IrisAnnealingEpsilonGreedy/AccuracyAftrerPrune.npy')
Softmax = np.load('/home/salm/Documents/svhn/IrisSoftmax/AccuracyAftrerPrune.npy')
AnnealingSoftmax = np.load('/home/salm/Documents/svhn/IrisAnnealingEpsilonGreedy/AccuracyAftrerPrune.npy')
Exp3 = np.load('/home/salm/Documents/svhn/IrisExp3/AccuracyAftrerPrune.npy')
Hedge = np.load('/home/salm/Documents/svhn/IrisHedge/AccuracyAftrerPrune.npy')
```

```python
fig = plt.figure(figsize=(10, 6), dpi=80)
ax = fig.add_subplot(111)
N = len(ucb1)
## necessary variables
ind = np.arange(N)                # the x locations for the groups
plt.plot(ind , ucb1 , color="blue", linewidth=2.5, linestyle="-", label="UCB1")
plt.plot(ind , EpsilonGreedy, color="red", linewidth=2.5, linestyle="-", label="Epsilon Greedy")
plt.plot(ind , AnnealingEpsilonGreedy, color="cyan", linewidth=2.5, linestyle="-", label="Annealing Epsilon Greedy")
plt.plot(ind , Softmax, color="magenta", linewidth=2.5, linestyle="-", label="Softmax")
plt.plot(ind , AnnealingSoftmax, color="yellow", linewidth=2.5, linestyle="-", label="Annealing Softmax")
plt.plot(ind , Exp3, color="black", linewidth=2.5, linestyle="-", label="Exp3")
plt.plot(ind , Hedge, color="green", linewidth=2.5, linestyle="-", label="Hedge")
plt.plot(ind , AccBefore, color="0.75", linewidth=2.5, linestyle="-", label="Acc Before Pruning")
plt.legend(loc = 3)
plt.axis([0, 400, 0, 1])
plt.grid(True)
plt.show()
```
